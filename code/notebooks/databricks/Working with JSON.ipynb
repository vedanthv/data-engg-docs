```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
restuarant_json_data = spark.read.format('json') \
    .option("nultiline",'true')\
        .option("inferSchema",'true')\
            .load("file:/Workspace/Users/vedanthvbaliga@gmail.com/Data/restuarant_data.json")
```

```python
display(restuarant_json_data)
```

```python
df_exploded_res = restuarant_json_data.select("*",explode("restaurants").alias('new_restaurants'))\
    .drop("restaurants")
display(df_exploded_res)
```

```python
df_exploded_res_1 = restuarant_json_data.select("*",explode("restaurants").alias('new_restaurants'))\
    .drop("restaurants")\
        .select('new_restaurants.restaurant.R.res_id')
display(df_exploded_res_1)
```

```python
df_exploded_res_new = restuarant_json_data.select("*",explode("restaurants").alias('new_restaurants'))\
    .drop("restaurants")\
        .select('*','new_restaurants.restaurant.R.res_id',explode_outer("new_restaurants.restaurant.establishment_types").alias('new_establishment_types'),"new_restaurants.restaurant.name")
display(df_exploded_res_new)
```
